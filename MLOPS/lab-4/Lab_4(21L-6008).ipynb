## Read the Dataset

```python
# 1) Load dataset
uci_url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/credit-screening/crx.data'
local_filename = 'crx.data'


if not os.path.exists(local_filename):
  try:
    print('Downloading dataset from UCI...')
    urllib.request.urlretrieve(uci_url, local_filename)
    print('Downloaded to', local_filename)
  except Exception as e:
    print('Could not download dataset automatically. Please download manually from:')
    print(uci_url)



# column names: A1..A15 and the last column is the class
col_names = [f'A{i}' for i in range(1, 16)] + ['class']


df = pd.read_csv(local_filename, header=None, names=col_names, na_values='?')
```

# Remove Missing Values

```python
# Placing median in the numeric column's NAN values
cols = df.select_dtypes(include=["int64","float64"])
print("Numeric Columns", cols.columns)
df[cols.columns] = cols.fillna(cols.median())

# Placing mod in the string column's NAN calues
cols_object = df.select_dtypes(include="object")
print("String Columns:", cols_object.columns)
df[cols_object.columns] = cols_object.fillna(cols_object.mode().iloc[0])

print("Total NA in whole df:",df.isna().sum().sum())
```

# Perform Label Encoding on String Columns

```python
# Apply LabelEncoder to each string column
print("\nWtihout Encoding\n",df[cols_object.columns].head())  # displaying only string values 

le = LabelEncoder()
for col in cols_object.columns:
    df[col] = le.fit_transform(df[col])
    
print("\nWith Encoding\n",df[cols_object.columns].head())  # displaying only string values 
```

# Remove Outliers form Data

```python
def check_normality(df, cols):
    for col in cols:
        data = df[col].dropna()
        print(f"\n--- {col} ---")
        
        # Plot histogram
        plt.figure(figsize=(4,4))
        plt.subplot(2,2,2)
        sns.histplot(data, kde=True)
        plt.title(f"Histogram of {col}")
        plt.tight_layout()
        plt.show()

# Run for numeric columns
check_normality(df, df.select_dtypes(include=["int64","float64"]))
```

```python
# As our dats is not normal we will use the IQR

# Example: remove outliers for all numeric columns
Q1 = df.quantile(0.25)   # 25th percentile
Q3 = df.quantile(0.75)   # 75th percentile
IQR = Q3 - Q1            # Interquartile range

outliers = (df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR))
# Outliers per column
print("Outliers in each column:\n",outliers.sum())
print("Original shape:", df.shape)
# Remove outliers
df_no_outliers = df[~outliers.any(axis=1)]
print("After removing outliers:", df_no_outliers.shape)
```

```python
import numpy as np

# Correlation matrix
corr_matrix = df.corr().abs()

# Upper triangle
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))

to_drop = []
for column in upper.columns:
    high_corr = upper[column][upper[column] > 0.5].index.tolist()
    if high_corr:
        for correlated in high_corr:
            print(f"Keeping: {column}, Dropping: {correlated}")
        to_drop.append(correlated)

# Drop them
df.drop(columns=to_drop, inplace=True)
print("After removing co-related columns:", df.shape)
```

# Find F1 Score, Precision, Recall and Confusion Matrix for each model

```python
results = []
for name, model in models.items():
    print('\n===== {} ====='.format(name))
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    print('Classification report:\n', classification_report(y_test, y_pred, target_names=le.classes_))
    cm = confusion_matrix(y_test, y_pred)
    print('Confusion matrix:\n', cm)
    acc = accuracy_score(y_test, y_pred)
    prec = precision_score(y_test, y_pred, average='macro', zero_division=0)
    rec = recall_score(y_test, y_pred, average='macro', zero_division=0)
    f1 = f1_score(y_test, y_pred, average='macro', zero_division=0)
    results.append({'model': name, 'accuracy': acc, 'precision_macro': prec, 'recall_macro': rec, 'f1_macro': f1})
```
