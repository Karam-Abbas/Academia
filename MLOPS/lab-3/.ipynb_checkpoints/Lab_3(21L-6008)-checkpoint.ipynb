## Task 1: Reading the Dataset
Read the dataset using Pandas with default indexing. 

```python
import pandas as pd
# Write your code below
df = pd.read_csv("data.csv")
```

## Step 2: Exploratory Data Analysis (EDA)
We check basic statistics and visualize missing values, distributions, etc.

```python
# Basic info
# Find the missing values
print("Missing value in each col:\n",df.isna().sum())
# Total Count of missing values in the whole dataframe
print("Total Missing Values",df.isna().sum().sum())
# Find the statistical data analysis on the data
df.describe()
```

## Step 3: Handling Missing Values
We can fill missing numeric values with median and categorical with mode.

```python
print("Total Columns",len(df.columns))
# drop columns with more than 70% of null values
df = df.loc[:, df.isna().mean() < 0.7]
print("Total Columns after Null Threshold",len(df.columns))

# Fill numeric columns with median
df.fillna(df.median(numeric_only=True), inplace=True)

# Fill categorical columns with mode
object_cols_mode = {col: df[col].mode()[0] for col in df.select_dtypes(include='object')}
df.fillna(object_cols_mode, inplace=True)

print("Total Missing Values", df.isna().sum().sum())
```

## Step 4: Encoding Categorical Variables
Convert categorical (string) values into numbers using Label Encoding and One-Hot Encoding.

```python
from sklearn.preprocessing import LabelEncoder

df_label = df.copy()
df_one_hot = df.copy()

cols = df.select_dtypes(include='object').columns
# Label Encoding
for col in cols:
    le = LabelEncoder()
    df_label[col] = le.fit_transform(df_label[col])

print("\nLabel Encoded DF:\n", df_label.shape)

# One-Hot Encoding
df_one_hot = pd.get_dummies(df_one_hot, columns=df_one_hot.select_dtypes(include='object').columns, drop_first=True)
print("\nOne-Hot Encoded DF:\n", df_one_hot.shape)
```

## Step 5: Outlier Detection
Use Z-scores to find outliers 

```python
import numpy as np
from scipy.stats import zscore
# use Z score to identify outliers display the rows having outliers but do not remove them for now.
# only have the numeric columns to apply z-score & IQR too
df_numeric = df.select_dtypes(include=['int64', 'float64'])
def detect_outliers_zscore(df, threshold=3):
    outliers = {}
    for col in df.columns:
        z_scores = np.abs(zscore(df[col].dropna()))
        outlier_idx = df[z_scores > threshold].index
        outliers[col] = outlier_idx
        print(f"{col}: {len(outlier_idx)}")
    return outliers
    
# Z-score calculation 
z_outliers = detect_outliers_zscore(df_numeric)
```

## Step 6: Outlier Detection and Removal (IQR)
We use IQR to detect and remove outliers.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Loop through each numeric column
for col in df_numeric.columns:
    plt.figure(figsize=(12,4))
    
    # Histogram
    plt.subplot(1,2,1)
    sns.histplot(df_numeric[col].dropna(), kde=True, bins=30)
    plt.title(f'Histogram of {col}')
    
    # Boxplot
    plt.subplot(1,2,2)
    sns.boxplot(x=df_numeric[col])
    plt.title(f'Boxplot of {col}')
    
    plt.show()
```

```python
# Use from scratch IQR method to identify and remove outliers. 
def remove_outliers_iqr(df):
    df_clean = df.copy()
    for col in df.select_dtypes(include=[np.number]).columns:
        Q1 = np.percentile(df[col].dropna(), 25)
        Q3 = np.percentile(df[col].dropna(), 75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        # outliers free df
        df_clean = df_clean[(df_clean[col] >= lower_bound) & (df_clean[col] <= upper_bound)]
    return df_clean

# Run
df_no_outliers = remove_outliers_iqr(df_numeric)

# Please make comparative study between z-score and IQR in identifying better method for identification of outliers on our dataset. 
# As per our data, better option is using the IQR to remove the outliers as our data is skewed and not normally distributed.

print("Original shape:", df_numeric.shape)
print("After removing outliers:", df_no_outliers.shape)
```

## Step 7: Feature Selection using Correlation
We check correlation with the target variable `SalePrice`. Variables with high correlation are more predictive.

```python
# Take SalePrice as a reference column and find 10 of the most highest correlated columns to price. 
correlation = df_numeric.corr()['SalePrice'].sort_values(ascending=False)

# Top 10 most correlated columns (excluding SalePrice itself)
top_corr_features = correlation.index[1:11]

print("Top 10 features correlated with SalePrice:\n")
print(correlation[1:11])

# For Machine Learning purposes and draw their heatmap. 
corr_matrix = df_numeric[top_corr_features.to_list() + ['SalePrice']].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Top 10 Correlated Features with SalePrice")
plt.show()
```

## Step 8: Dimensionality Reduction using PCA
We use Principal Component Analysis (PCA) to reduce dimensionality.

```python
# Reduce the dimensions of the data using PCA. Convert the columns into two prinicple component only. 
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# Standardize the dataset (So all the columns are compareable)
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df_numeric)

# Apply PCA (reduce to 2 components)
pca = PCA(n_components=2)
pca_data = pca.fit_transform(scaled_data)

# DataFrame with the 2 principal components
pca_df = pd.DataFrame(data=pca_data, columns=['PC1', 'PC2'])

# 4. Plot the 2 principal components
plt.figure(figsize=(8,6))
plt.scatter(pca_df['PC1'], pca_df['PC2'], alpha=0.5)
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA - Reduced to 2 Dimensions')
plt.show()
```
