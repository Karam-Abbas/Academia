```python
# q1
# for And gate

def activation(out, threshold):
    if out >= threshold:
        return 1
    else:
        return 0

def perceptron(a ,b):
    x1 = [0, 0, 1, 1]
    x2 = [0, 1, 0, 1]
    y = b
    w = [1, 1, 1]  # WEIGHTS
    threshold = a
    learning_rate = 0.2  
    bias = 1 
    i = 0
    print("Perceptron Training : \n")
    while i < 4:
        summation = bias * w[0] + x1[i] * w[1] + x2[i] * w[2]
        output = activation(summation, threshold)
        print("Input : " + str(x1[i]) + " , " + str(x2[i]))
        print("Bias : " + str(bias))
        print("Weights : " + str(w[0]) + " , " + str(w[1]) + " , " + str(w[2]))
        print("Summation : " + str(summation) + " Threshold : " + str(threshold))
        print("Actual Output : " + str(y[i]) + " Predicted Output : " + str(output))
        if output != y[i]:
            print("_______\nUpdating Weights")
            w[0] = w[0] + learning_rate * (y[i] - output) * bias
            w[1] = w[1] + learning_rate * (y[i] - output) * x1[i]
            w[2] = w[2] + learning_rate * (y[i] - output) * x2[i]
            print("Updated Weights : " + str(w[0]) + " , " + str(w[1]) + " , " + str(w[2]))
            i = -1
            print("\nWeights Updated Training Again : ")
            print("###################################")
        i += 1
        print("------------------")

    # Testing perceptron
    and_input = [1, 1]  # example for x1=1 and x2=1
    summation = bias * w[0] + and_input[0] * w[1] + and_input[1] * w[2]
    prediction = activation(summation, threshold)
    print("\nPrediction for input", and_input, ":", prediction)

print("And Gate")
# And gate threshold = 2
label = [0, 0, 0, 1]
threshold = 2
perceptron(threshold,label)
print("\n\n\n\n\n\n")
print("Or Gate")
# Or gate threshold = 1
label = [0, 1, 1, 1]
threshold = 1
perceptron(threshold,label)



```
